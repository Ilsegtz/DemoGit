# Futuros, Forwards y Opciones 

## FUTUROS
- Acuerdo para comprar o vender un activo en una fecha específica en el futuro a un precio determinado.
- Quien compara tiene una posción _LARGA_.
- Quien vende tiene una posción _CORTA_.
- El contrato se denomina de acuerdo con el mes de entrega. La posición corta decide cuando se hace la entrega. Por lo general, no termina en entrega. 
- La bolsa espicifica el lugar y el periodo extacto del mes para entrega.
- Existe un límite supererior en inferior de precios. 
- Límite de posiciones: Número máximo de contratos que el especulador puede mantener. 
- Debe tener una cuenta de margen. 
- Existen corredores a comision y locales.
- Algunos tipos de órdenes son: de mercado, limitada, de compra o venta, time-of-day. 

## FORWARDS
- Comprar o vender un activo en una fecha específica en el futuro a un precio determinado.
- Se negocian en el mercado _OTC_ (no inscritos en bolsa).
- Se hacenc otizaciones para transacciones grandes. 
- Los contratos sobre divisas son populares. 
- _Comerciantes al contado_: entraga inmediata.
- _Comerciantes al plazo_: entraga a futuro.

## OPCIONES
- _Opción de compra_: derecho de comprar un activo a cierto precio en una fecha específica.
- _Opción de venta_: derecho de vender un activo a cierto precio en una fecha específica.
- Precio ejercicio o _STRIKE_: es el precio establecido. 
- Opción _europea_: es al vencimiento. 
- Opción _americana_: ejerce en cualquier momento. 
- Al inicio se paga un prima de la opción para participar en contrato de opciones. 
- El precio de la opción de compra baja a medida de que sube el precio del ejercicio. 
- El mercado OTC de opciones puede adaptarse para satisfacer las necesidades particulares.
- Existen cuatro tipo de participantes en el mercado de opciones:
1. Compradores de opción de compra.
2. Vendedores de opción de compra.
3. Compradores de opción de venta.
4. Vendedores de opción de venta.
