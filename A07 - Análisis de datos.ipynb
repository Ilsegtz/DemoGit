Éstos corresponden a los contagios diarios de COVID-19 a partir del 22 de marzo del 2020 al 6 de agosto del 2021.

Ordene los datos en un DataFrame de pandas correspondiendo fecha con numero de contagios.

```python
#Fechas para comenzar y terminar
start="2020-03-22"
end="2021-08-06"
Fechas=pd.date_range(start,end)
```

Hacemos la tabla

```python
#Hacemos nuestras separacion
N=np.linspace(1,len(D),num=len(D))

#Hacemos la tabla
tabla = pd.DataFrame(columns=["Fecha","Datos"], index=N)

#Agregamos datos a la tabla
tabla["Fecha"] =Fechas
tabla["Datos"] = D

tabla
```

Calcule la variación relativa día a día que mejor se aproxime a una distrubución normal

```python
#Calculamos nuestros rendimientos
R = ((tabla["Datos"] - tabla["Datos"].shift())/tabla["Datos"].shift())
R=R.replace(np.inf,np.nan).dropna()
```

```python
# Asumimos que los rendimientos presentan una distribución normal
media= R.mean()
std= R.std()
media,std
```

```python
#Hacemos nuestra funcion para simular rendimientos
def simR(media, std, start, dias, N) :
    R_simulado = np.random.normal(media, std, size=(dias, N))
    fechas = pd.date_range(start, periods=dias)
    return pd.DataFrame(data=R_simulado, index=fechas)
```

```python
#Simulamos los rendimientos
rendimientos_simulados = simR(media, std,start, 365, 10000)
rendimientos_simulados
```

Simule 10000 escenarios para el siguiente año y calcule la probabilidad de que los contagios sean menores que 50.

```python
#Simulamos los rendimientos para el siguiente año
rendimientos_simulados2 = simR(media, std,"2022-01-01", 365, 10000)
```

```python
# Obtenemos los rendimientos acumulados
rendimientos_acumulados2 = (rendimientos_simulados2 + 1).cumprod()
```

```python
# Valores simulados
valores_simulados2 = tabla["Datos"].iloc[-1]*rendimientos_acumulados2
```

```python
#Calculamos la probabilidad
valor_deseado2 = tabla["Datos"].iloc[-1]*1.5 #Aqui va la probabilidad a calcular
sobrepasa_el_deseado = (valores_simulados2 < valor_deseado2) #Definimos que queremos calcular
probabilidad_de_sobrepasar = sobrepasa_el_deseado.mean(axis=1)
probabilidad_de_sobrepasar.plot(grid=True,figsize=(8, 5))
```
